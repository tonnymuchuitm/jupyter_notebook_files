```python
data = raw_data.drop(['Model'], axis = 1)
data.describe(include = 'all')
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data_no_null = data.dropna(axis=0)
```

```python
data_no_null.describe(include = 'all')
```

```python
sns.distplot(data_no_null['Price'])
```

```python
q = data_no_null['Price'].quantile(0.99)
cleaned_data = data_no_null[data_no_null['Price'] < q]
cleaned_data
```

```python
cleaned_data.describe(include = 'all')
```

```python
sns.distplot(cleaned_data['Price'])
```

```python
sns.distplot(cleaned_data['Mileage'])
```

```python
q = cleaned_data['Mileage'].quantile(0.99)
data_2 = cleaned_data[cleaned_data['Mileage']<q]
data_2
```

```python
data_2.describe(include = 'all')
```

```python
sns.distplot(data_2['Mileage'])
```

```python
sns.distplot(data_2['EngineV'])
```

```python
data_3 = data_2[data_2['EngineV']<6.5]
sns.distplot(data_3['EngineV'])
```

```python
sns.distplot(data_3['Year'])
```

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year']>q]
sns.distplot(data_4['Year'])
```

```python
datacleaned = data_4.reset_index(drop=True)
datacleaned.describe(include = 'all')
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize = (15,3))
ax1.scatter(datacleaned['Year'], datacleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(datacleaned['EngineV'],datacleaned['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(datacleaned['Mileage'],datacleaned['Price'])
ax3.set_title('Price and Mileage')


plt.show()
```
