## simple plotting example

this is a plot of the $sinc$ function

$$
f(x) = \frac{\sin(x)}{x}
$$

```python
%matplotlib inline

#this is a magic line in python that allows you to select your plottig backend, 
#there are a couple of plotting backends for matplotlib which are divided into two categories
#the first category is the non-interactive matplotlib backend
#this none interactive backends dont allow the use to interact with the plots eg changing the sizes of the plot to our desired size
#the most commonly used non-interactive matplotlib backend is the %matplotlib inline
#this inline backend stores your plot together with your code

#the second is the interactive matplotlib backend 
#the most common of this interactive backend being the %matplotlib notebook 
#interactive backends allow the user to interact with the plot eg
#allowing you to either execute your code on an existing plot or create a new plot to execute your new code
#they also allow you to configure the plot to whichever size you want
#this does not automatically store your plot but gives you an interactive plot that has an option to store.
#you can store your plot in whichever form you prefer and at a selected folder of your choice which makes this backend 
#a little bit preffered by jupyter notebook users
#it is worth noting that the magic line should be written before you import your matplotlib module
#it is not advisable to change the backend midway through your project. it can be fatal


import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

plt.figure(figsize=(6,4))
pi = np.pi#we are setting our pi as np.pi
x = np.linspace(-4*pi, 4*pi, 1000)#linspace call allows you to select a range in which you want your points to be in and the number of points you want in between that range
#the first parameter is the least point of your range and the second parameter is the maaximum point of your range, the third parameter is the number of points you want in between the range

plt.plot(x, np.sin(x)/x, label='sin(x) curve', color='red')#plt.plot takes in the x-axis and the second argument is the y-axis
plt.axhline(0, color= 'black', lw=1)#plt.axhline creates a horizontal origin line and takes in the first parameter as the y-axis at which to plot the horizontal line
plt.axvline(0, color='black', lw=1)#plt.axvline creates a vertical origin line and takes in the first parameter as the x-axis at which to place the vertical line
plt.xlabel('pie')
plt.ylabel('f(x)=sin(x)/x')
plt.xticks([-4*pi,-2*pi, 0, 2*pi, 4*pi], ['$-4\pi$', '$-2\pi$', '0', '$2\pi$','$4\pi$'])#plt.xticks takes in two parameters, the first being the list of arguments and the second being the list of keyword arguments
plt.yticks([-0.5, 0, 0.5, 1],['-0.5', '0', '0.5', '1.0'])
plt.legend(loc=2, title='legend', fontsize=6)
plt.show()
```
