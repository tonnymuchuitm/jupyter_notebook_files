```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
#a histogram is a barchart which shows the frequency of a certain phenomena
#simply a histogram is a barchart where the x-axis is a given observation and the y-axis is the frequency with which that observation occurs
```

```python
#how to plot a single hitogram in matplotlib
#a histogram only needs a one-dimension array to plot
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins=3, rwidth=0.98)#by default matplotlib sets your bins at 10 but you can change it to any number of your preference
#we also have to specify the width of our data, by default matplotlib will set the width of our bars as 1, that means our bars will be cojoined which is not a good representation.
#
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.98, color = 'grey')#by default matplotlib will set the range of our bin according to the number of bins we will pass but we can specify that by editing the bins parameter to pass in a list of the ranges we want.
#we can also pass in the color we want for our histogram
```

```python
bloodsugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
bloodsugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]

#we can also plot two histograms side by side. we only need to pass in our different data as a list
#and also pass in their different colors as a list.
plt.hist([bloodsugar_men, bloodsugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green', 'orange'], label=['men', 'women'], orientation = 'vertical')
plt.legend()
plt.xlabel('blood sugar')
plt.ylabel('frequency')
plt.title('blood sugar by gender')
```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n = {}'.format(sample_size))
```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample, bins=100)
    axs[n].set_title('n = {}'.format(sample_size))
```

```python
# use gridspec to partition the figure into subplots
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 1:])
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])
```

```python
Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
lower_right.scatter(X, Y)
top_histogram.hist(X, bins=100)
s = side_histogram.hist(Y, bins=100, orientation='horizontal')
```

```python
# clear the histograms and plot normed histograms
top_histogram.clear()
top_histogram.hist(X, bins=100, density=True)
side_histogram.clear()
side_histogram.hist(Y, bins=100, orientation='horizontal', density=True)
# flip the side histogram's x axis
side_histogram.invert_xaxis()
```

```python
# change axes limits
for ax in [top_histogram, lower_right]:
    ax.set_xlim(0, 1)
for ax in [side_histogram, lower_right]:
    ax.set_ylim(-5, 5)
```
