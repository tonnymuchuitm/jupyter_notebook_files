```python
import numpy as np
x = np.array([1,2,3,4,5,6,7,8])#our x-axis points
y = x#our y-axis points which are the same as our x-axis
colours = ['green']*(len(x)-1)#assigning green colour to all our points except the last point of our graph
colours.append('red')#we assign the last point of our graph as red in colour

plt.figure()#this command lets us have a new figure after our code is executed, if not passed our code will get excecuted on our previous figure.
plt.scatter(x,y,s=100, c=colours)#plt.scatter takes in the x-axis as the first parameter, y-axis as the second parameter , 
#s which is scaler as the third parameter and c which is colour as the last parameter
```

## Lineplots

```python
import numpy as np

linear_data = np.array([1,2,3,4,5,6,7,8])
quadratic_data = linear_data**2

plt.figure()
plt.plot(linear_data,'-o', quadratic_data, '-o')#takes in a parameter accompanied by an appearance of a plot eg -x or -o, the - before the appearance is critical because it informs python that we want a link line
#first we only gave only our y-axis to our plot call, no x-axis values,
# Instead, the plot function was smart enough to figure out that what we wanted was to use the index of the series as the x value. Which is pretty handy when you want to make quick plots.
#Second we see that the plot identifies this as two series of data and that the colors of the data from the series are different including the data points and the lines between the data points. This is different from the scatter plot which required us to label the lines directly.
```

```python
plt.xlabel('some data')
plt.ylabel('some other data')
plt.title('A title')
plt.legend(['Baseline', 'Competition'])#since we did not labelthe data when plotting we have to pass in a list of the naming legend in a similar way as we passed in our data
#And we can create a legend too. But note that since we didn't label the data points as we did with the scatter plot, we need to create legend entries when we add the legend itself.
```

```python
plt.gca().fill_between(range(len(linear_data)),#first parameter is the x-axis
                      linear_data, quadratic_data,#second parameter the y1, third parameter the y2
                      facecolor = 'green')#fourth parameter is the colour we want to fill in 
```

```python
plt.figure()

observation_dates = np.arange('2017-01-01', '2017-01-09', dtype = 'datetime64[D]')
plt.plot(observation_dates, linear_data, '-o',
        observation_dates, quadratic_data, '-o')
```

## Removing all the ticks on all our axis except the x-axis

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

plt.bar(pos, popularity, align='center')
plt.xticks(pos, languages, rotation=60)
plt.ylabel('% Popularity')
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top=False, bottom=True, left=False, right=False, labelleft=False, labelbottom=True)
plt.show()
```

## Removing the frames of a chart

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

plt.bar(pos, popularity, align='center')
plt.xticks(pos, languages)
plt.ylabel('% Popularity')
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=True)

# remove the frame of the chart
for spine in plt.gca().spines.values():#we have to iterate thru all spines.values which are our four sides of the graph
    spine.set_visible(False,)#and for each spine we set it as invisible
plt.show()#we run plt.show() to show our graph without any frames
```

## Differently colouring the bars of the chart

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# change the bar colors to be less bright blue
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')
# make one bar, the python bar, a contrasting color
bars[0].set_color('blue')

# soften all labels by turning grey
plt.xticks(pos, languages, alpha=0.8)
plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=True)

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.show()
```

## Directly labbeling the bars of the chart and letting go off the y-axis

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# change the bar color to be less bright blue
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')
# make one bar, the python bar, a contrasting color
bars[0].set_color('#1F77B4')

# soften all labels by turning grey
plt.xticks(pos, languages, alpha=0.8)
# remove the Y label since bars are directly labeled
#plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=True)

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
# direct label each bar with Y axis values
for bar in bars:
    plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() - 5, str(int(bar.get_height())) + '%', 
                 ha='center', color='w', fontsize=11)
plt.show()
```
